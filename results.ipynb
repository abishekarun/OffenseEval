## SubTask B

```
import numpy as np

def get_report_taskb(cm):
    recall = np.diag(cm) / np.sum(cm, axis = 1)
    precision = np.diag(cm) / np.sum(cm, axis = 0)

    def get_f1score(p,r):
      return 2*p*r/(p+r)

    f1_score = get_f1score(precision,recall)
    
    print('The precision for TIN is: {:.5f}'.format(precision[0]))
    print('The recall for TIN is: {:.5f}'.format(recall[0]))
    print('The f1 score for TIN is: {:.5f}'.format(f1_score[0]))

    print('The precision for UNT is: {:.5f}'.format(precision[1]))
    print('The recall for UNT is: {:.5f}'.format(recall[1]))
    print('The f1 score for UNT is: {:.5f}'.format(f1_score[1]))

    print('The average precision is: {:.5f}'.format(np.mean(precision)))
    print('The average recall is: {:.5f}'.format(np.mean(recall)))
    print('The macro average f1 score is: {:.5f}'.format(np.mean(f1_score)))
```

### Dev Set

```
# XGBoost 0.61248
cm = [[742,33],[81,24]]
get_report_taskb(cm)
```

```
# Ensemble 0.61231
cm = [[693,82],[69,36]]
get_report_taskb(cm)
```

```
# Synthetic Ensemble 0.6175
cm = [[729,77],[46,28]]
get_report_taskb(cm)
```
